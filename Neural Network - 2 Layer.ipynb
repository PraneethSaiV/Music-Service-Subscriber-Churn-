## Neural Network Architecture

```python
x = tf.placeholder(tf.float32, [None, 104])
y_ = tf.placeholder(tf.float32, [None,2])
```

```python
W1 = tf.Variable(tf.truncated_normal([104,300], stddev=0.1))
b1 = tf.Variable(tf.constant(0.1, shape=[300]))
```

```python
W2 = tf.Variable(tf.truncated_normal([300,2], stddev=0.1))
b2 = tf.Variable(tf.constant(0.1, shape=[2]))
```

```python
y = tf.matmul((tf.matmul(x,W1) + b1),W2) + b2
```

```python
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_, logits=y))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
prediction = tf.nn.softmax(y)
```

## Start Session

```python
def obtain(dataset):
    return dataset.drop('is_churn', axis = 1).values , pd.get_dummies(pd.DataFrame(dataset.is_churn),columns=['is_churn']).values
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
batch_size = 10000
```

```python
for _ in range(1000):
    if _ % 100 == 0:
        print(_)
    batch_data = machinedf_train.sample(batch_size)
    x_set, y_set = obtain(batch_data)
    feed_dict = {x: x_set,
                y_: y_set}
    train_step.run(feed_dict = feed_dict, session=sess)
```
